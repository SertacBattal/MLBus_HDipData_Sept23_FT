### K-Means & DBSCAN

```python
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt
```

```python
# KMeans ile kmeleme
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_selected_data)
```

```python
# Grselletirme
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(scaled_selected_data[:, 0], scaled_selected_data[:, 1], c=kmeans_labels, cmap='viridis')
plt.title('KMeans Clustering')
plt.xlabel('MIN (Minutes)')
plt.ylabel('PTS (Points)')
```

```python
from sklearn.metrics import silhouette_score

db_param_options = [[0.1, 5], [0.15, 7], [1.0, 15], [0.05, 7], [0.25, 5], [0.35, 9]]
for ep,min_sample in db_param_options:
    
    # Generate clusters using DBSCAN
    db = DBSCAN(eps=ep, min_samples = min_sample)
    db_clusters = db.fit_predict(scaled_selected_data)
    
    # Check if only one cluster is found
    if len(np.unique(db_clusters)) > 1:
        print("Eps: ", ep, "Min Samples: ", min_sample)
        print("DBSCAN Clustering: ", silhouette_score(scaled_selected_data, db_clusters))
    else:
        print("Eps: ", ep, "Min Samples: ", min_sample)
        print("DBSCAN Clustering: Only one cluster found, silhouette score cannot be computed.")
```

```python
# DBSCAN ile kmeleme
dbscan = DBSCAN(eps=0.35, min_samples=9)
dbscan_labels = dbscan.fit_predict(scaled_selected_data)


plt.subplot(1, 2, 2)
plt.scatter(scaled_selected_data[:, 0], scaled_selected_data[:, 1], c=dbscan_labels, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('MIN (Minutes)')
plt.ylabel('PTS (Points)')

plt.tight_layout()
plt.show()
```

```python
from sklearn_extra.cluster import KMedoids
from sklearn.cluster import AgglomerativeClustering

# K-medoids ile kmeleme
kmedoids = KMedoids(n_clusters=3, random_state=42)
kmedoids_labels = kmedoids.fit_predict(scaled_selected_data)

# Hiyerarik Kmeleme ile kmeleme
hierarchical = AgglomerativeClustering(n_clusters=3)
hierarchical_labels = hierarchical.fit_predict(scaled_selected_data)

# Grselletirme
plt.figure(figsize=(12, 6))

# K-medoids
plt.subplot(1, 2, 1)
plt.scatter(scaled_selected_data[:, 0], scaled_selected_data[:, 1], c=kmedoids_labels, cmap='viridis')
plt.title('KMedoids Clustering')
plt.xlabel('MIN (Minutes)')
plt.ylabel('PTS (Points)')

# Hiyerarik Kmeleme
plt.subplot(1, 2, 2)
plt.scatter(scaled_selected_data[:, 0], scaled_selected_data[:, 1], c=hierarchical_labels, cmap='viridis')
plt.title('Hierarchical Clustering')
plt.xlabel('MIN (Minutes)')
plt.ylabel('PTS (Points)')

plt.tight_layout()
plt.show()
```

```python
# Kmeleme algoritmalarn ve deerlendirme metriklerini aklama
"""
Kmeleme Algoritmalar:
- KMeans: Veri setinde belirli bir sayda kme oluturur. KMeans algoritmas basit ve hzldr.
- DBSCAN: Younluk tabanl kmeleme yntemidir. Veri setinde farkl ekillerdeki kmeleri tanmlar.
- KMedoids: KMeans'e benzer, ancak medyanlar kullanr, bu nedenle daha dayankldr.
- Hiyerarik Kmeleme: Kme saysn nceden belirlemeden kmeleme yapar.

Deerlendirme Metrikleri:
- Siluet Skoru: Kmeleme sonularnn homojenliini ve ayrmlln deerlendirir. +1'e yaklaan deerler daha iyi bir kmeleme iaret eder.
- Davies-Bouldin ndeksi: Kmeleme sonularnn homojenlii ve ayrmll arasndaki dengeyi deerlendirir. Dk deerler daha iyi bir kmeleme iaret eder.
"""

# Kmeleme algoritmalar ve deerlendirme metriklerini uygulama
from sklearn_extra.cluster import KMedoids
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score

# KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_selected_data)
silhouette_kmeans = silhouette_score(scaled_selected_data, kmeans_labels)
db_kmeans = davies_bouldin_score(scaled_selected_data, kmeans_labels)

# DBSCAN
dbscan = DBSCAN(eps=0.35, min_samples=9)
dbscan_labels = dbscan.fit_predict(scaled_selected_data)
silhouette_dbscan = silhouette_score(scaled_selected_data, dbscan_labels)
db_dbscan = davies_bouldin_score(scaled_selected_data, dbscan_labels)

# KMedoids
kmedoids = KMedoids(n_clusters=3, random_state=42)
kmedoids_labels = kmedoids.fit_predict(scaled_selected_data)
silhouette_kmedoids = silhouette_score(scaled_selected_data, kmedoids_labels)
db_kmedoids = davies_bouldin_score(scaled_selected_data, kmedoids_labels)

# Hiyerarik Kmeleme
hierarchical = AgglomerativeClustering(n_clusters=3)
hierarchical_labels = hierarchical.fit_predict(scaled_selected_data)
silhouette_hierarchical = silhouette_score(scaled_selected_data, hierarchical_labels)
db_hierarchical = davies_bouldin_score(scaled_selected_data, hierarchical_labels)

# Sonularn karlatrlmas
print("KMeans - Siluet Skoru: {:.3f} // Davies-Bouldin ndeksi: {:.3f}".format(silhouette_kmeans, db_kmeans))
print("DBSCAN - Siluet Skoru: {:.3f} // Davies-Bouldin ndeksi: {:.3f}".format(silhouette_dbscan, db_dbscan))
print("KMedoids - Siluet Skoru: {:.3f} // Davies-Bouldin ndeksi: {:.3f}".format(silhouette_kmedoids, db_kmedoids))
print("Hiyerarik Kmeleme - Siluet Skoru: {:.3f} // Davies-Bouldin ndeksi: {:.3f}".format(silhouette_hierarchical, db_hierarchical))

```

Hangi kmeleme algoritmalarnn tercih edilecei ve neden?:
Sonulara gre, siluet skoru ve Davies-Bouldin indeksi deerlerine dayanarak, Hiyerarik Kmeleme ve KMeans algoritmalarnn daha iyi performans gsterdii grlyor. Bu algoritmalar, daha yksek siluet skorlarna ve daha dk Davies-Bouldin indekslerine sahiptir. Bu nedenle, veri setinde bu algoritmalardan biri tercih edilebilir.
Siluet skoru, bir kmelemenin homojenliini ve ayrmlln ler. 1'e yaklaan bir siluet skoru, kmelerin birbirinden iyi ayrld ve birbirlerine benzeyen veri noktalarnn ayn kmeye atanmas durumunun minimum olduu anlamna gelir. Davies-Bouldin indeksi ise, kmeleme sonularnn homojenlii ve ayrmll arasndaki dengeyi deerlendirir. Dk bir deer, daha homojen ve daha iyi ayrlm kmeleri gsterir.
Sonularn Yorumlanmas:
Hiyerarik Kmeleme ve KMeans algoritmalar, daha yksek siluet skorlarna ve daha dk Davies-Bouldin indekslerine sahiptir, bu da daha homojen ve daha iyi ayrlm kmeler elde edildiini gsterir.
DBSCAN ve KMedoids algoritmalar, daha dk siluet skorlar ve daha yksek Davies-Bouldin indeksleri gsterir. Bu, bu algoritmalarn belirli veri seti iin daha uygun olmayabileceini ve daha az homojen veya ayrlm kmeler oluturabileceini gsterebilir.
Dolaysyla, segmentasyon iin tercih edilen algoritmalar genellikle daha yksek siluet skoru ve daha dk Davies-Bouldin indeksi deerlerine sahip olanlardr. Bu durumda, Hiyerarik Kmeleme ve KMeans algoritmalar tercih edilebilir.

```python
# import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf 
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm
import statsmodels.tsa.api as smtsa
```

```python
# plotds is method to plot time series, ACF and PACF
def plotds(xt, nlag = 30, fig_size = (12, 10)):
    if not isinstance(xt, pd.Series):
         xt = pd.Series(xt)
    plt.figure(figsize=fig_size)
    layout = (2, 2)
    
    ax_xt = plt.subplot2grid(layout, (0, 0), colspan = 2)
    ax_acf= plt.subplot2grid(layout, (1, 0))
    ax_pacf = plt.subplot2grid(layout, (1, 1))
    
    xt.plot(ax = ax_xt)
    ax_xt.set_title('Time Series')
    plot_acf(xt, lags = 50, ax = ax_acf)
    plot_pacf(xt, lags = 50, ax = ax_pacf)
    plt.tight_layout()
    return None
```

```python
# Example to show the difference between ACF and PAF
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# We generate a synthetic dataset representing daily temperature readings over a year.
np.random.seed(0)
days = np.arange(1, 366)
temperature = 20 + np.sin(2 * np.pi * days / 365) * 10 + np.random.normal(0, 3, 365)

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# We plot the temperature data to visualize the underlying pattern.
axs[0].plot(days, temperature)
axs[0].set_title('Daily Temperature Readings')
axs[0].set_xlabel('Day')
axs[0].set_ylabel('Temperature (C)')
axs[0].grid(True)

# The ACF plot shows the correlation between the temperature values at different lags. For example, we might observe 
# high autocorrelation at lag 1 (temperature today vs. temperature yesterday), indicating that consecutive days are correlated.
plot_acf(temperature, lags=40, ax=axs[1])
axs[1].set_title('Autocorrelation Function (ACF)')
axs[1].set_xlabel('Lag')
axs[1].set_ylabel('Autocorrelation')
axs[1].grid(True)

# The PACF plot, on the other hand, shows the partial autocorrelation, which represents the correlation between two time points 
# after removing the effects of the intervening observations. It helps identify the direct relationships between observations 
# at different lags.
plot_pacf(temperature, lags=40, ax=axs[2])
axs[2].set_title('Partial Autocorrelation Function (PACF)')
axs[2].set_xlabel('Lag')
axs[2].set_ylabel('Partial Autocorrelation')
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

ARIMA Model Parameter Determination:

The appropriate parameters (p, d, q) for the ARIMA model were determined through a combination of visual analysis and statistical methods. The Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) plots were utilized to identify the values of p and q, which represent the number of lag observations included in the model. The parameter d, representing the degree of differencing, was determined based on the stationarity of the time series data.
Evaluation of ARIMA Model Performance:

The ARIMA model was fitted to the closing price of the stock data, and its performance was evaluated using mean squared error (MSE) as the evaluation metric. The model demonstrated relatively good performance in capturing the overall trend and variability in the closing prices, as indicated by the low MSE value.
Additionally, future values of the closing price were forecasted using the trained ARIMA model. The forecasted values were compared against the actual closing prices to assess the model's accuracy in predicting future trends. While the model provided reasonable forecasts for the immediate future, its performance may deteriorate for longer forecast horizons due to the inherent uncertainty in financial markets.
Strengths and Limitations of the ARIMA Model:

Strengths: The ARIMA model is capable of capturing both trend and seasonality in time series data, making it suitable for forecasting stock prices, which often exhibit such patterns. Additionally, the model's simplicity and interpretability make it easy to implement and understand.
Limitations: One limitation of the ARIMA model is its assumption of linear relationships between variables, which may not hold true in all cases, particularly in highly volatile and nonlinear financial markets. Moreover, the model's reliance on historical data for forecasting future values may lead to inaccuracies, especially during periods of significant market disruptions or regime shifts.
